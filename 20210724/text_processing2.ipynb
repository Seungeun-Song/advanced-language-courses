```python
def nGrams(doc_string_or_list, n, docIsAlreadTokenized):
    doc = doc_string_or_list
    if not docIsAlreadTokenized:
        doc = [token.text for token in nlp(doc_string_or_list)]
    doc = " ".join(doc).lower().split(' ')
    grams = [doc[i:i+n] for i in range(len(doc)-n+1)]
    return grams
```

```python
doc = "How old are you today OR can you tell me somthing about you"
n = 4

# n = 1 : 유니그램(unigram)
# n = 2 : 바이그램(bigram)
# n = 3 : 트라이그램 (trigram)
# n >= 4 : n-gram

grams = nGrams(doc, n, False)
```

```python
def oneHotVector(word, W2I, numWords):
    v = np.zeros(numWords)
    v[W2I[word]] = 1
    return v
```

```python
doc = "How are you today. I know most of the time how you feel. ."

# 이 doc를 토큰화한 다음 리스트로 만들어보시오.

tokens = [token.text for token in nlp(doc)]
numWords = len(tokens)
print(tokens, "\n", numWords, "\n", type(tokens))

W2I = dict(zip(tokens, np.arange(numWords)))
I2W = dict(zip(np.arange(numWords), tokens))

print(W2I["feel"])
print(I2W[13])
```
