```python
import nltk
from nltk.corpus import webtext

webtext_words = webtext.words("firefox.txt")
word_frequency = nltk.FreqDist(webtext_words)

from nltk.corpus import stopwords

stop_words = stopwords.words("english")


type(word_frequency)
```

```python
word_frequency
```

```python
webtext_without_stopwords = dict([ 
    (k,v) for k, v in word_frequency.items() if k not in stop_words and len(k) > 1
])

print(len(webtext_without_stopwords), type(webtext_without_stopwords))
print(webtext_without_stopwords)
```

```python
frequency_dist = nltk.FreqDist(webtext_without_stopwords)
frequency_dist.plot(10)
```

```python
from wordcloud import WordCloud

wcloud = WordCloud().generate_from_frequencies(frequency_dist)


import matplotlib.pyplot as plt
plt.imshow(wcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
